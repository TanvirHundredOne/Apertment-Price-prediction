```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
X1=pd.read_csv('train.csv',index_col='Id')
X2=pd.read_csv('test.csv',index_col='Id')
Y1=X1.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X=X1[features].copy()
X_test=X2[features].copy()
X_train,X_valid,y_train,y_valid=train_test_split(X,Y1,train_size=0.8,test_size=0.2,random_state=19)
```

```python
X_test.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Define the models
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='mae', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth=7, random_state=0)

models = [model_1, model_2, model_3, model_4, model_5]
```

```python
from sklearn.metrics import mean_absolute_error
for i in range(0,len(models)):
    model=models[i]
    X_t=X_train
    X_v=X_valid
    y_t=y_train
    y_v=y_valid
    model.fit(X_t,y_t)
    prediction=model.predict(X_v)
    mae=mean_absolute_error(y_v,prediction)
    print(mae)
```

```python
model[2].fit(X,Y1)
preds_test=model[2].predict(X_test)
output = pd.DataFrame({'Id': X_test.index,
                       'SalePrice': preds_test})
print(output)
output.to_csv('output.csv', index=False)
```
